## PCA로 feats 차원축소해서 그려보기

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(feats_r)
```

```python
pca_result.shape
```

```python
pca_res = DataFrame()
pca_res['pca-one'] = pca_result[:,0]
pca_res['pca-two'] = pca_result[:,1]

print( 'Explained variation per principle component: {}'.format(pca.explained_variance_ratio_))
```

```python
pca_res['pca-label'] = label_r_res
pca_res['pca-label-no'] = [np.argmax(label_r_no.iloc[i]) for i in range(len(label_r_no))]
```

```python
pca_res 
```

```python
pca_res['pca-one'][0:10]
```

```python
pca_res.groupby('pca-label').size()
```

```python
plt.scatter(pca_res['pca-one'],pca_res['pca-two'],c=pca_res['pca-label-no'],s=1)

plt.show() ## 색깔별 뭔지 주석을 좀 달아야함
```

```python
groups = pca_res.groupby('pca-label')

fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group['pca-one'], 
            group['pca-two'], 
            marker='o', 
            linestyle='',
            markersize=2,
            label=name)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
          fancybox=True, shadow=True, ncol=5) # legend position

plt.title('Scatter Plot', fontsize=20)
plt.xlabel('PCA-One', fontsize=14)
plt.ylabel('PCA-Two', fontsize=14)

plt.show()
```

```python
colorr.values
```

```python
stne_grp = DataFrame()
stne_grp['xs'] = xs
stne_grp['ys'] = ys
stne_grp['label'] = pca_res['pca-label']
stne_grp['label-no'] = pca_res['pca-label-no']
```

```python
stne_grps = stne_grp.groupby('label')

fig, ax = plt.subplots()
for name, group in stne_grps:
    ax.plot(group['xs'], 
            group['ys'], 
            marker='.', 
            linestyle='',
            label=name,
            markersize=3
           )

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
          fancybox=True, shadow=True, ncol=5) # legend position


plt.title('Scatter Plot', fontsize=20)
plt.xlabel('xs', fontsize=14)
plt.ylabel('ys', fontsize=14)

plt.show()
```
